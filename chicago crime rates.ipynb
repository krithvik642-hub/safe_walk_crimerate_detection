```python
import pandas as pd
import numpy as np

# Load data (adjust path if needed)
df = pd.read_csv("/Users/akumaresan/Downloads/archive/crimes.csv", low_memory=False)

# Drop unnecessary columns (as you started earlier, plus more for simplicity)
df = df.drop(columns=[
    'Unnamed: 0', 'ID', 'Case Number', 'Block', 'IUCR', 'FBI Code', 
    'X Coordinate', 'Y Coordinate', 'Location', 'Updated On'
], errors='ignore')

# Convert 'Date' to datetime (handle mixed formats)
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y %I:%M:%S %p', errors='coerce')

# Drop rows where Date couldn't be parsed (rare, but safe)
df = df.dropna(['Date'])

# Extract time-based features for ML
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['Hour'] = df['Date'].dt.hour
df['Weekday'] = df['Date'].dt.weekday  # 0=Monday, 6=Sunday

# Handle missing values
# Categoricals: fill with 'Unknown'
categorical_cols = ['Primary Type', 'Description', 'Location Description', 'District', 'Ward', 'Community Area']
for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].fillna('Unknown')

# Numerics: impute with median (e.g., Beat, Latitude, Longitude)
numeric_cols = ['Beat', 'Latitude', 'Longitude']
for col in numeric_cols:
    if col in df.columns:
        df[col] = df[col].fillna(df[col].median())

# Booleans are already clean (Arrest, Domestic)

# Remove duplicates
df = df.drop_duplicates()

# Optional: Filter to recent years for faster processing (e.g., 2010+)
df = df[df['Year'] >= 2010]

# Check cleaned info
df.info()
print(f"Cleaned shape: {df.shape}")
df.head()
```

```python
# ─── 1. Import only the tools we really need ───
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

# ─── 2. Prepare monthly crime numbers ───
# Group by month → count how many crimes happened each month
monthly = df.groupby(pd.Grouper(key='Date', freq='M')).size().reset_index(name='Crimes')

# Add useful columns for the model to learn patterns
monthly['Month'] = monthly['Date'].dt.month          # January = 1, December = 12
monthly['Year']  = monthly['Date'].dt.year
monthly['Crimes_last_month'] = monthly['Crimes'].shift(1)     # what happened last month
monthly['Crimes_last_year']  = monthly['Crimes'].shift(12)    # same month last year

# Remove first rows that have missing "previous" values
monthly = monthly.dropna()

# ─── 3. Choose what the model should use to guess future crimes ───
X = monthly[['Month', 'Year', 'Crimes_last_month', 'Crimes_last_year']]
y = monthly['Crimes']

# ─── 4. Split: learn from past, test on recent months (important for time data!) ───
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# ─── 5. Create and train a simple smart guesser (Random Forest) ───
model = RandomForestRegressor(n_estimators=80, random_state=42)
model.fit(X_train, y_train)

# ─── 6. Check how good the guesses are ───
predictions = model.predict(X_test)

error = mean_absolute_error(y_test, predictions)
print(f"Average error: about ± {error:.0f} crimes per month")
print(f"Lower number = better model")

# ─── 7. Show picture: real vs model's guess ───
plt.figure(figsize=(10, 5))
plt.plot(monthly['Date'].iloc[-len(y_test):], y_test, label='Real crimes', color='blue', linewidth=2)
plt.plot(monthly['Date'].iloc[-len(y_test):], predictions, label="Model's guess", color='orange', linestyle='--', linewidth=2)
plt.title("Real vs Predicted Monthly Crimes (last part of data)")
plt.xlabel("Date")
plt.ylabel("Number of crimes")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Prepare one row for next month
last_row = monthly.iloc[-1:]

next_month = last_row.copy()
next_month['Date'] = last_row['Date'] + pd.offsets.MonthEnd(1)
next_month['Month'] = next_month['Date'].dt.month
next_month['Year']  = next_month['Date'].dt.year
# Use last real values as starting point for lags
next_month['Crimes_last_month'] = last_row['Crimes']
next_month['Crimes_last_year']  = monthly.iloc[-12]['Crimes'] if len(monthly) >= 12 else last_row['Crimes']

print("Predicted crimes next month:", int(model.predict(next_month[['Month','Year','Crimes_last_month','Crimes_last_year']])[0]))
```

```python
import streamlit as st
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

# Page setup - modern look
st.set_page_config(
    page_title="Chicago Crimes Interactive Dashboard",
    page_icon="🚨",
    layout="wide"
)

st.title("🚨 Chicago Crimes Explorer")
st.markdown("Interactive dashboard for Chicago crime data (based on your dataset)")

# ────────────────────────────────────────────────
# Load data with caching
# ────────────────────────────────────────────────
@st.cache_data(show_spinner="Loading your Chicago crimes data…")
def load_crimes():
    path = "/Users/akumaresan/Downloads/archive/crimes.csv"
    try:
        df = pd.read_csv(path, low_memory=False)
    except FileNotFoundError:
        st.error(f"File not found: {path}\nPlease check the path is correct.")
        st.stop()
    
    # Quick cleaning
    if 'Date' in df.columns:
        df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
        df['Year'] = df['Date'].dt.year
        df['Month'] = df['Date'].dt.month_name()
        df['DayOfWeek'] = df['Date'].dt.day_name()
        df['Hour'] = df['Date'].dt.hour
    
    # Drop useless column if present
    if 'Unnamed: 0' in df.columns:
        df = df.drop(columns=['Unnamed: 0'])
    
    return df

df = load_crimes()

# Show basic info
with st.expander("Dataset Overview"):
    st.write(f"Rows: {len(df):,}")
    st.write(f"Columns: {list(df.columns)}")
    st.dataframe(df.describe(), use_container_width=True)

# ────────────────────────────────────────────────
# Sidebar Filters
# ────────────────────────────────────────────────
st.sidebar.header("Filters")

# Year range
if 'Year' in df.columns:
    min_year = int(df['Year'].min())
    max_year = int(df['Year'].max())
    selected_years = st.sidebar.slider(
        "Year Range",
        min_year, max_year,
        (max_year - 5, max_year)
    )

# Primary Type (if exists)
if 'Primary Type' in df.columns:
    crime_types = sorted(df['Primary Type'].dropna().unique())
    selected_types = st.sidebar.multiselect(
        "Crime Types",
        options=crime_types,
        default=crime_types[:10]
    )

# District / Beat / Ward / Community Area
for col in ['District', 'Beat', 'Ward', 'Community Area']:
    if col in df.columns:
        values = sorted(df[col].dropna().unique().astype(str))
        selected = st.sidebar.multiselect(f"{col}", values)
        if selected:
            df = df[df[col].astype(str).isin(selected)]

# Apply year & type filters
filtered = df.copy()

if 'Year' in df.columns:
    filtered = filtered[
        (filtered['Year'] >= selected_years[0]) &
        (filtered['Year'] <= selected_years[1])
    ]

if 'Primary Type' in df.columns and selected_types:
    filtered = filtered[filtered['Primary Type'].isin(selected_types)]

# ────────────────────────────────────────────────
# Key Metrics Cards
# ────────────────────────────────────────────────
st.subheader("Quick Stats")
cols = st.columns(4)

with cols[0]:
    st.metric("Total Incidents", f"{len(filtered):,}")
with cols[1]:
    st.metric("Years Covered", f"{filtered['Year'].min()} – {filtered['Year'].max()}" if 'Year' in filtered else "N/A")
with cols[2]:
    if 'Primary Type' in filtered.columns:
        st.metric("Most Common Crime", filtered['Primary Type'].mode()[0] if not filtered.empty else "N/A")
with cols[3]:
    st.metric("Unique Locations", filtered['Community Area'].nunique() if 'Community Area' in filtered else "N/A")

# ────────────────────────────────────────────────
# Tabs for different views
# ────────────────────────────────────────────────
tab_map, tab_trends, tab_top = st.tabs(["🗺️ Map", "📈 Trends", "🏆 Top Crimes"])

with tab_map:
    st.subheader("Crime Locations")
    if 'Latitude' in filtered.columns and 'Longitude' in filtered.columns:
        # Scatter map
        fig_scatter = px.scatter_mapbox(
            filtered,
            lat="Latitude",
            lon="Longitude",
            color="Primary Type" if 'Primary Type' in filtered else None,
            hover_data=['Year', 'District', 'Beat'],
            zoom=10,
            height=600,
            opacity=0.5
        )
        fig_scatter.update_layout(mapbox_style="carto-positron")
        st.plotly_chart(fig_scatter, use_container_width=True)

        # Density heatmap
        st.markdown("**Crime Density Heatmap**")
        fig_heat = px.density_mapbox(
            filtered,
            lat="Latitude",
            lon="Longitude",
            radius=15,
            zoom=10,
            height=500
        )
        fig_heat.update_layout(mapbox_style="carto-positron")
        st.plotly_chart(fig_heat, use_container_width=True)
    else:
        st.info("No Latitude/Longitude columns found → map disabled")

with tab_trends:
    st.subheader("Trends Over Time")
    if 'Year' in filtered.columns:
        yearly = filtered.groupby('Year').size().reset_index(name='Count')
        fig_year = px.line(yearly, x='Year', y='Count', title="Crimes per Year", markers=True)
        st.plotly_chart(fig_year, use_container_width=True)

    if 'DayOfWeek' in filtered.columns:
        dow = filtered['DayOfWeek'].value_counts().reindex(
            ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
        ).reset_index()
        fig_dow = px.bar(dow, x='DayOfWeek', y='count', title="Crimes by Day of Week")
        st.plotly_chart(fig_dow, use_container_width=True)

with tab_top:
    st.subheader("Top Crime Types")
    if 'Primary Type' in filtered.columns:
        top_n = st.slider("Show top N types", 5, 20, 10)
        top_crimes = filtered['Primary Type'].value_counts().head(top_n).reset_index()
        top_crimes.columns = ['Crime Type', 'Count']
        
        fig_top = px.bar(
            top_crimes,
            x='Count',
            y='Crime Type',
            orientation='h',
            title=f"Top {top_n} Crime Types",
            color='Count',
            color_continuous_scale='Reds'
        )
        fig_top.update_layout(yaxis={'categoryorder':'total ascending'})
        st.plotly_chart(fig_top, use_container_width=True)
    else:
        st.info("No 'Primary Type' column found")

# Raw data preview
with st.expander("View Filtered Data"):
    st.dataframe(filtered.head(2000))

st.caption("Data from: /Users/akumaresan/Downloads/archive/crimes.csv")
```
